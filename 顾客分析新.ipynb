```python
df1=pd.read_csv('processed1.csv')
```

```python
df1=pd.get_dummies(df1, columns=['Age','CityType','Profession','Sex','YearsInCity'])
```

```python
df1.describe()
```

```python
def scale_minmax(col):
    return (col-col.min())/(col.max()-col.min())

scale_cols = []
for i in range(0,18):
    scale_cols.append('a'+str(i))

df1[scale_cols] = df1[scale_cols].apply(scale_minmax,axis=0)
df1[scale_cols].describe()
```

```python
df1.head()
```

```python
test=df1[df1["Married"].isnull()]
train=df1[df1["Married"].notnull()]
train_y=train["Married"].astype(int)
train_X=train.drop(['Married','CustomerID'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
train_X, valid_X, train_y, valid_y = train_test_split(train_X, train_y, test_size=0.1,random_state=4)
```

```python
train_X.head()
```

```python
test_X = test.drop(['Married','CustomerID'], axis = 1)
test_X.head()
```
